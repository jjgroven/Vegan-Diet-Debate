```python
#Import packages
import pandas as pd
import numpy as np
```

```python
#Read Data
data = pd.read_csv("CV_MaxFeatures.csv")
data
```

```python

# Convert BoW matrix to Transaction Format
transactions = []

for _, row in data.iterrows():
    # Select column names where value is greater than 0 (present words)
    transaction = [col for col in data.columns if row[col] > 0] # Extract words present in document
    transactions.append(transaction)
```

```python
transactions = pd.DataFrame(transactions)
transactions.to_csv("transactionData.csv",index = False)
```
